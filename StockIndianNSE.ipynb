# Retrieving stock quotes from NSE exchange for Indian stocks

```python
import quandl
```

```python
import matplotlib.pyplot as plt
```

```python
#mydata = quandl.get('NSE/SBIN', start_date="2017-1-1", end_date="2017-12-31")
mydata = quandl.get('NSE/SBIN', start_date="2017-1-1")
```

```python
print(mydata.head())
```

```python
print(mydata.tail())
```

```python
print(mydata.info())
```

```python
mydata['Close'].plot()
```

```python
plt.show()
```

```python
## get the volatility chart

stock_return = mydata.apply(lambda x: x / x[0])
```

```python
print(stock_return.head())
```

```python
stock_return['Close'].plot(title='Return')
```

```python
plt.show()
```

```python
min(stock_return['Close'])
```
